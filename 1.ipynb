```python
import requests
from bs4 import BeautifulSoup
import time
import csv
```

```python
url = "https://quotes.toscrape.com/"
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
quotes = soup.find_all('div', class_='quote')
```

```python
for quote in quotes:
    text = quote.find('span', class_='text').text
    author = quote.find('small', class_='author').text
    print(f'{text} - {author}')
```

```python
page_url = "https://quotes.toscrape.com/page/"
```

```python
for page in range(1, 6):
    print(f"Scraping page {page}")

    response_page = requests.get(page_url + str(page))
    soup_page = BeautifulSoup(response_page.text, 'html.parser')
    quotes_page = soup_page.find_all('div', class_='quote')
    for quote in quotes_page:
        text = quote.find('span', class_='text').text
        author = quote.find('small', class_='author').text
        print(f'{text} - {author}')
    time.sleep(2)
```

```python
current_url = "https://quotes.toscrape.com/page/1/"
filename = "quotes_data.csv"
page_count = 1

with open(filename, "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(["Quote", "Author"])

    while current_url:
        print(f"Page = {page_count}. Scraping Url: {current_url}")
        response_current = requests.get(current_url)
        soup_current = BeautifulSoup(response_current.text, 'html.parser')
        quotes_current = soup_current.find_all('div', class_='quote')
        for quote in quotes_current:
            text = quote.find('span', class_='text').text
            author = quote.find('small', class_='author').text
            writer.writerow([text, author])

        #getting to next page
        next_button = soup_current.find('li', class_='next')
        if next_button:
            current_url = "https://quotes.toscrape.com" + next_button.find('a')['href']
            time.sleep(2)
        else:
            current_url = None
        page_count += 1

print("Scraping Done")
```
